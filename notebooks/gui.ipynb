```python
import os
import gradio as gr
import pandas as pd
from pycaret.regression import load_model  # Use the appropriate model type
# from pycaret.classification import load_model  # Uncomment for classification

# Load the trained PyCaret model
MODEL_PATH = "best_model"  # Replace with your actual model file path or name

model = load_model(MODEL_PATH)  # Load the trained model


# Function to make predictions using the loaded model
def make_prediction(*feature_values):
    try:
        # Map feature values to their respective names
        features = {name: value for name, value in zip(feature_names, feature_values)}

        # Convert input features into a DataFrame (PyCaret requires a DataFrame for prediction)
        feature_data = pd.DataFrame([features])

        # Perform prediction
        prediction = model.predict(feature_data)

        # Return the predicted value
        return f"Predicted Value: {prediction.tolist()[0]}"
    except Exception as e:
        return f"An error occurred: {str(e)}"


# Read the feature names dynamically from a CSV file
CSV_FILE = "file.csv"  # Replace with the path to your CSV file
TARGET_COLUMN = "TARGET"  # Replace with the name of the target column

if not os.path.exists(CSV_FILE):
    raise FileNotFoundError(f"CSV file not found at {CSV_FILE}")

# Load feature names from the CSV file (excluding the target column)
df = pd.read_csv(CSV_FILE)
feature_names = [col for col in df.columns if col != TARGET_COLUMN]


# Create Gradio interface
def create_gradio_interface():
    # Dynamically create input fields for features
    input_fields = [
        gr.Textbox(label=f"{feature}", placeholder=f"Enter value for {feature}") for feature in feature_names
    ]

    # Output field for prediction
    output_field = gr.Textbox(label="Prediction Output", interactive=False)

    # Interface layout
    interface = gr.Interface(
        fn=make_prediction,
        inputs=input_fields,
        outputs=output_field,
        title="GU",
        description="Enter feature values to make predictions using a PyCaret-trained model.",
    )

    return interface


# Run the Gradio interface
if __name__ == "__main__":
    gr_interface = create_gradio_interface()
    gr_interface.launch()
```
