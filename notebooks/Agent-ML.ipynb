```python
import pandas as pd
from pycaret.regression import setup, compare_models, pull, save_model
# from pycaret.classification import setup, compare_models, pull, save_model

# Load CSV data
def load_data(file_path):
    return pd.read_csv(file_path)

# Perform the PyCaret ML workflow with a list of algorithms
def pycaret_workflow(file_path, target_column, algorithms=None):
    # Step 1: Load the data
    data = load_data(file_path)

    # Step 2: Setup PyCaret environment
    clf_setup = setup(data, target=target_column,session_id=123)

    # Step 3: Compare models using a list of specified algorithms
    algorithms = ['en', 'br', 'rf', 'gbr', 'xgboost', 'lightgbm']
    if algorithms:
        best_model = compare_models(include=algorithms)
    else:
        best_model = compare_models()

    # Step 4: Retrieve and print the model comparison results
    results = pull()
    print("Model Comparison Results:\n", results)

    # Step 5: Save the best model
    save_model(best_model, 'best_model')

    print("\nThe best model has been saved as 'best_model.pkl'")

# Replace 'your-data.csv' with the path to your CSV file
# Replace 'target' with the name of the target column in your data

# Specify a list of algorithms to compare
# algorithms = ['rf', 'xgboost', 'lightgbm',  'svm']

data_path = r"C:\Users\Tejas\Downloads\ML Dataset\Student_Marks.csv"
Target = "Marks"

# pycaret_workflow('/content/sample_data/california_housing_train.csv', 'median_house_value', algorithms)
pycaret_workflow(data_path, Target, )
```
